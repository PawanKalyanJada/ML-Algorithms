```python
import random
class Kmeans:
    def __init__(self,n=3):
        self.n=n
        
    def distance(self,x,x1):
        a=sum([(x[i]-x1[i])**2 for i in range(len(x))])
        return a**0.5
        
    def fit(self,x):
        self.best_labels=None
        self.best={}
        var=float('inf')
        for i in range(10):
            self.labels=[None]*x.shape[0]
            d={}
            for i in range(self.n):
                d[i]=random.choice(x)
            while(True):
                d1={}
                c=[0]*self.n
                for j in range(x.shape[0]):
                    l=[]
                    for k in d:
                        l.append([self.distance(x[j],d[k]),k])
                    l.sort(key=lambda p:p[0])
                    self.labels[j]=l[0][1]
                for j in range(self.n):
                    d1[j]=np.array([0.0]*x.shape[1])
                for j in range(x.shape[0]):
                    v=self.labels[j]
                    d1[v]+=x[j]
                    c[v]+=1
                for j in range(self.n):
                    if(c[j]!=0):
                        d1[j]/=c[j]
                    else:
                        print(d1[j],c[j])
                b1=0
                for j in d:
                    if((d[j]==d1[j]).sum()==len(d[j])):
                        b1+=1
                if(b1==self.n):
                    break
                else:
                    d=d1
            var1=0
            for j in range(x.shape[0]):
                v=self.labels[j]
                var1+=self.distance(x[j],d[v])
            if(var1<var):
                self.best=d 
                self.best_labels=self.labels
                var=var1
```

```python
m=Kmeans(n=3)
m.fit(df.values)
```
